```python
import pandas as pd
import numpy as np
```

```python
cols = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration", "num_doors", "body_style", "drive_wheels", "engine_location", "wheel_base", "length", "width", "height", "curb_weight", "engine_type", "num_cylinders", "engine_size","fuel_system","bore", "stroke", "compression_ratio", "horsepower", "peak_rpm","city_mpg", "highway_mpg", "price"]
```

```python
car_prices = pd.read_csv("imports-85.data", names=cols)
```

```python
car_prices.head(5)
```

```python
numerical_cols = ["normalized_losses", "wheel_base", "length", "width", "height", "curb_weight", "engine_size", "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm", "city_mpg", "highway_mpg", "price"]
```

```python
numerical_car_prices = car_prices[numerical_cols]
```

```python
#replace all the ? values with numpy.nan
numerical_car_prices.replace("?", np.nan)
```

```python
numerical_car_prices = numerical_car_prices.replace("?", np.nan)
```

```python
numerical_car_prices = numerical_car_prices.astype(dtype=float, copy=False)
```

```python
normalized_losses = numerical_car_prices["normalized_losses"]
normalized_losses_null = pd.isnull(normalized_losses)
len(normalized_losses[normalized_losses_null])
```

```python
normalized_losses_mean = np.mean(normalized_losses[normalized_losses_null==False])
numerical_car_prices[normalized_losses_null]["normalized_losses"] = normalized_losses_mean
                
```
