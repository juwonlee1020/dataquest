```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("AmesHousing.txt", delimiter = "\t")
```

```python
train = data[0:1460]
test = data[1460:]
```

```python
train_null_counts = train.isnull().sum()
#dataframe of no missing value: selecting columns that contain no missing values
df_no_mv = train[train_null_counts[train_null_counts==0].index]
print(df_no_mv.head(10))
print(df_no_mv.isnull().sum())
```

### Converting non-numerical values to categorical values

```python
#choosing categorical columns
#text_cols = df_no_mv.select_dtypes(include=['object']).columns
text_cols = df_no_mv.select_dtypes(include=['object']).columns
for col in text_cols:
    df_no_mv[col] = df_no_mv[col].astype('category')
    print(col)
    print(df_no_mv[col].unique())
    print(df_no_mv[col].cat.codes.value_counts())
    print("------------------------------------")
```

### Dummy coding
Instead of having a single column with n possible integer codes, we have n <br> binary columns

```python
dummy_cols = pd.DataFrame()
for col in text_cols:
    dummy_cols = pd.get_dummies(train[col])
    train = pd.concat([train, dummy_cols], axis=1)
    del train[col]
```
